## Plot

```python
def my_plot(s):
    fig, ax = plt.subplots(3, 2, constrained_layout=True, )
    fig.suptitle(str(s.year.iloc[0]), fontsize=20)
    ax[0,0].plot(s.YM, s.ITEM_CNT)
    ax[0,0].set_title("ITEM_CNT")
    ax[0,1].plot(s.YM, s.PRICE)
    ax[0,1].set_title("PRICE")
    ax[1,0].plot(s.YM, s.MAXTEMP)
    ax[1,0].set_title("MAX_TEMP")
    ax[1,1].plot(s.YM, s.SALEDAY)
    ax[1,1].set_title("SALEDAY")
    ax[2,0].plot(s.YM, s.RAIN_DAY)
    ax[2,0].set_title("RAINDAY")
    ax[2,1].plot(s.YM, s.HOLIDAY)
    ax[2,1].set_title("HOLYDAY")
    fig.set_figheight(10)
    fig.set_figwidth(10)
    plt.figure(figsize=(5.5,14))
    plt.subplot(414)
    plt.plot(s.YM, s.QTY)
    plt.title("QTY")
```

### Define Model Function

```python
def apply_model(x, y, x_test, y_test, model='lm'):
    
    if model == 'lm':
        m = LinearRegression()
        m.fit(x, y)
        
    elif model =='dt':
        m = DecisionTreeRegressor()
        m.fit(x, y)
        
    elif model =='svm':
        m = SVC()
        m.fit(x, y)
        
    elif model =='rf':
        m = RandomForestRegressor()
        m.fit(x, y)
        
    elif model == 'ann':
        
        m = Sequential()
        
        m.add(Dense(16, input_dim=7))
        m.add(Activation('relu'))
        m.add(Dropout(0.2))
        m.add(Dense(8))
        m.add(Activation('relu'))
        m.add(Dropout(0.2))
        m.add(Dense(1, activation='relu'))
        
        m.summary()
        
        m.complie(loss='mse', 
                  optimzier='', 
                  metrics=['accuracy'])
        
        m.fit(x, y, batch_size=10, epochs=10, validation_data=(x_test, y_test))
        
    else:
        raise ExceptionError("ML model is not correct!, plz confirm the type of model")
        
    predictions = m.predict(x_test)
    y_real = np.array(y_test)
    rmse = math.sqrt(mean_squared_error(y_real, predictions))
    
    return m, predictions, rmse
```
